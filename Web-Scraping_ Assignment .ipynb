# 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
# import the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
```

```python
page
```

```python
soup= BeautifulSoup(page.content)

soup
```

# Scraping Names

```python
names =[] # empty list for store the

for i in soup.find_all('div',class_="presidentListing"):
    names.append(i.text)
    
names
```

# Scraping Terms of office

```python
terms =[]# empty list

for i in soup.find_all('span',class_="terms"):
    terms.append(i.text)
    
terms
```

```python
# Printing length
print(len(names),len(terms))
```

```python
#Making dataframe
import pandas as pd
df = pd.DataFrame({'names':names,'terms':terms})
df
```

# 6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-

i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
# import the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
```

```python
page
```

```python
soup= BeautifulSoup(page.content)
soup
```

# Scraping Multipal Titles

```python
titles =[]# empty list for store the

for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    titles.append(i.text)
    
titles
```

# Scraping Multipal Authors

```python
Authors =[]# empty list

for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    Authors.append(i.text)
    
Authors
```

# Scraping the Published Date

```python
Published =[]# empty

for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    Published.append(i.text)
    
Published
```

# Scraping the Paper URL

```python
Paper =[]# empty

for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    Paper.append(i.get('sc-1thf9ly-0 sc-1thf9ly-1 iwnLUR fXmEge'))
    
Paper
```

```python
# Printing length
print(len(titles),len(Authors),len(Published),len(Paper))
```

```python
#Making dataframe
import pandas as pd
df = pd.DataFrame({'titles':titles,'Authors':Authors,'Published':Published,'Paper':Paper})
df
```
